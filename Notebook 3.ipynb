# 🔵 UNIVARIATE ANALYSIS

## 📌 Distribution of Sales Amount

```python
plt.figure(figsize=(6,4))
sns.histplot(df['sales_amount'], kde=True)
plt.title("Distribution of Sales Amount")
plt.show()
```

Insight:
This plot shows how sales values are distributed. A right-skewed distribution indicates that a small number of transactions contribute to very high sales.


## `📌 Distribution of Profit`

```python
plt.figure(figsize=(6,4))
sns.histplot(df['profit'], kde=True)
plt.title("Distribution of Profit")
plt.show()
```

Insight:
This helps identify whether there are loss-making transactions (negative values) and how profit is spread across orders.



## 📌 Product Category Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='product_category', data=df)
plt.title("Product Category Distribution")
plt.xticks(rotation=45)
plt.show()
```

Insight:
Shows the frequency of each product category, helping identify the most commonly sold items.



## 📌 Region Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='region', data=df)
plt.title("Region Distribution")
plt.show()
```

Insight:
Displays which regions generate the highest number of transactions.



# 🔗 BIVARIATE ANALYSIS

## 📌 Sales vs Profit

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='sales_amount', y='profit', data=df)
plt.title("Sales vs Profit")
plt.show()
```

Insight:
This visualization checks whether higher sales lead to higher profit or if there are inconsistencies due to costs or discounts.


## 📌 Sales by Region

```python
plt.figure(figsize=(6,4))
sns.barplot(x='region', y='sales_amount', data=df)
plt.title("Sales by Region")
plt.show()
```

Insight:
Highlights which regions contribute the most to total sales.


## 📌 Profit by Product Category

```python
plt.figure(figsize=(6,4))
sns.barplot(x='product_category', y='profit', data=df)
plt.title("Profit by Product Category")
plt.xticks(rotation=45)
plt.show()
```

Insight:
Identifies categories that are highly profitable versus those that may be generating losses.

## 📌 Discount vs Profit

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='discount_percent', y='profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

Insight:
Shows how discount levels impact profitability. Higher discounts often reduce profit margins.


# 🔥 MULTIVARIATE ANALYSIS




## 📌 Sales by Region and Product Category

```python
plt.figure(figsize=(8,5))
sns.barplot(x='region', y='sales_amount', hue='product_category', data=df)
plt.title("Sales by Region and Product Category")
plt.show()
```

Insight:
Provides a comparative view of how different product categories perform across regions.


## 📌 Correlation Heatmap

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Insight:
Displays relationships between numerical variables. Strong correlations indicate meaningful relationships.


## 📌 Pairplot

```python
sns.pairplot(df.select_dtypes(include=['number']))
plt.show()
```

Insight:
Offers a comprehensive visualization of relationships between all numerical features.
