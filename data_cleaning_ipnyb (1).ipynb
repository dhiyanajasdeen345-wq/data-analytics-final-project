## 🛠️ Step 5: Handling Missing Values

We apply different strategies based on the type of data:

### 🔹 Categorical Data
Missing values are replaced with **"Unknown"**

### 🔹 Numerical Data
- Sales_Amount → filled with mean  
- Discount_Percent → filled with median  
- Customer_Rating → filled with median  

This approach maintains data balance without introducing bias.

```python
import pandas as pd

df = pd.read_excel('Sales_ecommerce_data set (1).xlsx')

df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
df['Product_Category'] = df['Product_Category'].fillna('Unknown')
df['Payment_Method'] = df['Payment_Method'].fillna('Unknown')

df['Sales_Amount'] = df['Sales_Amount'].fillna(df['Sales_Amount'].mean())
df['Discount_Percent'] = df['Discount_Percent'].fillna(df['Discount_Percent'].median())
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].median())
```

```python
df.isnull().sum()
```

### ✅ Result

All missing values have been handled successfully.  
The dataset is now complete and consistent.
